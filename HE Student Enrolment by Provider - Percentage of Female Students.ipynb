## HE Student Enrolment by Provider - Percentage of Female Students

```python
import pandas as pd
```

## HE Student Enrolments by Provider - Full-time 2020/1

```python
df = pd.read_excel('HE student enrolments by HE provider 2020-1 FT.xlsx', index_col = [1], skiprows=17)
df.head()
```

```python
df
```

There are some 'Not known' values in the dataset relating to gender, which we wouldn't expect. There are three gender categories; female, male and other. As there are only 35 not known values in the dataset, we will exclude these from the analysis as they will be outliers that skew our analysis.

```python
# Drop 'Not known' column

df.drop('Not known', axis='columns',inplace=True)
df
```

```python
# Drop 'Not known' column as we will perform our own calculations

df.drop('UKPRN', axis='columns',inplace=True)
df
```

```python
df['Percentage of Female Students'] = (df['Female'] / df['Total'] * 100)
```

```python
df
```

```python
df['Percentage of Female Students'] = df['Percentage of Female Students'].round(2)
```

```python
df.sort_values(by=['Percentage of Female Students'],ascending=False, inplace=True )
df
```

```python
top10_2020 = df.head(10)
```

```python
top10_2020
```

```python
top10_2020.to_csv('top_10_providers_by_percentage_of_females_2020-1')
```

```python
top10_2020['Percentage of Female Students']
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (15,10))
top10_2020['Percentage of Female Students'].plot.bar()
plt.title("Top Ten HE Providers for Full-Time Female Enrolled Students by Percentage 2020/1")
plt.xlabel("HE provider")
plt.ylabel("Percentage of Female Students")
plt.grid(True, linewidth = 1,  linestyle = "--")
plt.show()
```

```python
plt.savefig('topten%2020-1.png')
```

```python
df.tail(10)
```

```python
bottom10_2020 = df.tail(10)
```

```python
bottom10_2020
```

```python
bottom10_2020.to_csv('bottom_10_providers_by_percentage_of_females_2020-1')
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (15,10))
bottom10_2020['Percentage of Female Students'].plot.bar()
plt.title("Bottom Ten HE Providers for Full-Time Female Enrolled Students by Percentage 2020/1")
plt.xlabel("HE provider")
plt.ylabel("Percentage of Female Students")
plt.grid(True, linewidth = 1,  linestyle = "--")
plt.show()
```

```python
plt.savefig('bottomten%2020-1.png')
```
