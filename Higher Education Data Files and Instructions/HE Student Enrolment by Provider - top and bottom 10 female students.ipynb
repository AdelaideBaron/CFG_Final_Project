## HE Student Enrolments by Provider - Full-time 2020/1

HE student enrolments by Provider 2020/1
Data downloaded from https://www.hesa.ac.uk/data-and-analysis/students/table-1 

Filters used= 

- Country of HE provider = England 
- Region of HE provider = All 
- First year marker = First year 
- Level of study = All 
- Mode of study = Full-time 
- Academic year = 2020/21 


This is all student enrolments for the academic year 2020/1

```python
df = pd.read_excel('HE student enrolments by HE provider 2020-1 FT.xlsx', index_col = [1], skiprows=17)
df.head()
```

UKPRN is the UK Provider Reference Number and is a unique identifier for each higher education institution.

```python
df.head()
```

```python
df
```

There are some 'Not known' values in the dataset relating to gender, which we wouldn't expect. There are three gender categories; female, male and other. As there are only 35 not known values in the dataset, we will exclude these from the analysis as they will be outliers that skew our analysis.

```python
df.info()
```

There are 237 rows and 6 columns in the dataset. The HE provider column is object data type as it is names of providers and all other columns are integers as they are counts of students.

There are 237 HE providers in our dataset, each of them a unique value.

```python
df.describe()
```

```python
# Drop 'Not known' column

df.drop('Not known', axis='columns',inplace=True)
df
```

```python
# Drop 'Not known' column as we will perform our own calculations

df.drop('Total', axis='columns',inplace=True)
df
```

```python
# Drop 'Not known' column as we will perform our own calculations

df.drop('UKPRN', axis='columns',inplace=True)
df
```

```python
df.to_csv('HEP_FT-2020_1')
```

```python
df = pd.read_csv('HEP_FT-2020_1', index_col = [0])
df
```

```python
df.sort_values(by=['Female'],ascending=False, inplace=True )
df
```

```python
update_df = df.drop('Total', inplace=True)
```

```python
highest_no_of_females = df.head(10)
highest_no_of_females
```

```python
highest_no_of_females.to_csv('top_ten_females_HEP_FT-2020_1')
```

```python
highest_no_of_females['Female']
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (15,10))
highest_no_of_females['Female'].plot.bar()
plt.title("Top Ten HE Providers for Full-Time Female Enrolled Students 2020/1")
plt.xlabel("HE provider")
plt.ylabel("Number of Female Students Enrolled")
plt.grid(True, linewidth = 1,  linestyle = "--")
plt.show()
```

```python
plt.savefig('toptenFT20201.png')
```

```python
lowest_no_of_females = df.tail(10)
lowest_no_of_females
```

There are 7 HE providers who did not haev any female enrolled students in 2020/1.
