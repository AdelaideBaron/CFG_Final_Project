# HE student enrolments by subject of study

```python
import pandas as pd
```

## HE student enrolments by subject of study FT 2019-0

Data downloaded from https://www.hesa.ac.uk/data-and-analysis/students/table-46
Filters used= Academic Year - 2019/0
Mode of Study = Full-time
Level of study = All (Undergraduate, Postgraduate Taught and Postgraduate Research students)

This is all student enrolments for the academic year 2019/0

```python
df = pd.read_excel("HE student enrolments by subject of study FT 2019-0.xlsx", index_col = 0, skiprows = 15)
df.head()
```

```python
df
```

Note the 'Not Known' column, with a value of 0 for every row. This row can be dropped from the data set with a caveat that they are all zeros. Not known is not a category we would expect to see in gender so it can be safely removed from the dataframe.

```python
df.info()
```

The dataframe has 25 rows corresponding to the CAH Level 1 subject group and 5 columns.The subject has been set as the index column. All other columns have a data type of integer, as they are counts of students. 

Whilst each CAH group has a number (i.e. Medicine and Dentistry is 01), there are numbers missing in the sequence. There is no 08 or 012, which is due to review of subject groupings.

There are also two total rows, one for Total science CAH level 1 and the other for Total non-science CAH level 1. These rows will need to be removed when counting the individiual subjects to avoid double counting.

```python
df.describe()
```

There are no missing values in the dataframe. All number are integers and are counts of students in each grouping.

```python
df['Female']
#This table shows the number of females studying each subject full-time in 2019/0.
```

```python
update_df = df.drop( 'Total non-science CAH level 1', inplace=True)
```

```python
update_df = df.drop( 'Total science CAH level 1', inplace=True)
```

```python
update_df = df.drop( 'Total', inplace=True)
```

```python
import matplotlib.pyplot as plt
```

```python
# drop Not known column
df.drop('Not known', axis='columns',inplace=True)
df
```

```python
df.to_csv('FT_enrol_2019_20')
```

```python
df[['Female']].plot.bar()
plt.title("Count of Females per CAH Level 1 Subject - Full-time 2019/0")
plt.xlabel("CAH Level 1 Subject")
plt.ylabel("Number of enrolled Females")
plt.grid(True, linewidth = 1,  linestyle = "--")
plt.show()
```

```python
df[['Female','Male', 'Other']].plot.bar()
plt.title("Students enroled per CAH Level 1 Subject by gender - Full-time 2019/0")
plt.xlabel("CAH Level 1 Subject")
plt.ylabel("Number of enrolled Students")
plt.grid(True, linewidth = 1,  linestyle = "--")
plt.show()
```
