# Graduate Salaries by Gender

```python
import pandas as pd
```

## Percentage of graduates in full-time paid employment in the UK by salary band and personal characteristics 2019/0

Data downloaded from https://www.hesa.ac.uk/data-and-analysis/sb263/figure-13

Filters used:

- Country of provider = England 
- Domicile = All 
- Provider type = Higher education providers (HEPs) 
- Level of qualification = All 
- Mode of former study = All 
- Work population marker = Paid employment is an activity 
- Academic year = 2019/20 


This is graduate outcomes for the academic year 2019/0

```python
df = pd.read_excel('Percentage of graduates in full-time paid employment in the UK by salary band and personal characteristics 2019-20.xlsx', index_col=0, skiprows = 17)
df
```

This is a small dataset and there are no null or not know values in the data. Data is only available for male female gender split for the salary information. Note percentages for females and males do not sum to 100%. This is because not all students are in paid employment, they may be undertaking further study for example.

```python
df.info()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
```

```python
df.plot.bar()
plt.title("Graduate Salaries by Gender 2019/0")
plt.xlabel("Salary Band")
plt.ylabel("Percentage of Graduates")
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.grid(True, linewidth = 1,  linestyle = "--")
plt.show()
```

```python
plt.savefig('grad_salaries_2019-0.png')
```

To compare the salarie bands for females for the 3 years, we need to merge the datasets.

```python
df4 = df.merge(df2, on='Salary band')
```

```python
df4
```

```python
female_salaries = df4.merge(df3, on='Salary band')
```

```python
female_salaries 
```

```python
# Drop 'Male' columns

female_salaries .drop('Male_x', axis='columns',inplace=True)
female_salaries .drop('Male_y', axis='columns',inplace=True)
female_salaries .drop('Male', axis='columns',inplace=True)
female_salaries
```

```python
female_salaries.rename(columns = {'Female_x':'% Females 2019/0', 'Female_y':'% Females 2018/9', 'Female':'% Females 2017/8'}, inplace=True)
```

```python
female_salaries
```

```python
female_salaries.to_csv('Female_graduae_salaries_2017-8_to_2019-0')
```

```python
female_salaries.plot.bar()
plt.title("Female Graduate Salaries 2017/8 to 2019/0")
plt.xlabel("Salary Band")
plt.ylabel("Percentage of Female Graduates")
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.grid(True, linewidth = 1,  linestyle = "--")
plt.show()
```
