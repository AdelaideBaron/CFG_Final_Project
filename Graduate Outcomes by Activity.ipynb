```python
import pandas as pd
```

## Graduate Outcomes by Activity Undergraduate Full-time 2019/20

Graduate outcomes by activity and personal characteristics

Data downloaded from https://www.hesa.ac.uk/data-and-analysis/sb263/figure-5 

Filters used:
- Country of provider = England 
- Domicile = All 
- Provider type = Higher education providers (HEPs) 
- Level of qualification = Undergraduate 
- Mode of former study = Full-time 
- Interim study = Include significant interim study 
- Academic year	2019/20 

This is graduate outcomes for the academic year 2019/0


```python
df = pd.read_excel("UG FT Grad outcomes 2019-20.xlsx", index_col=0, skiprows = 17)
df.head()
```

```python
df
```

There are NaN's in the dataframe but they are in rows we do not require for our analysis. These are in the category header rows, e.g. 'Sex', 'Disability status'. We are only looking at Gender and therefore only need the rows where the personal characteristic is 'Female', 'Male' or 'Other'. 

The totals columns can be removed from the dataset as we will be performing our own calculations. The columns 'Total with known outcomes' and 'Total' can be removed. 

We will also drop the non-respondents column as these are graduates that did not respond to the survey. We do not know what they are doing post degree completion and their inclusion will skew our data.

```python
df.drop('Total with known outcomes', axis='columns',inplace=True)
```

```python
df.drop('Total', axis='columns',inplace=True)
```

```python
df.drop('Non-respondents', axis='columns',inplace=True)
```

```python
df.rename(columns = {'Full-timeemployment':'Full-time Employment', 'Part-timeemployment':'Part-time Employment','Unknown pattern ofemployment':'Unknown Pattern of Employment', 'Employmentand further study':'Employment and Further Study'}, inplace=True)
```

```python
df
```

```python
df.to_csv('UG_grad_outcomes_2019-0')
```

```python
df_gender = df.head(4) 
df_gender
```

```python
df_chart = df_gender.drop(df.index[[0]])
df_chart
```

We will drop the 'Other' category for gender as the numbers are so small they will skew our analysis. We will focus on the activities for female and male graduates.

```python
df_chart.info()
```

There are now 3 rows in our dataframe for each of the gender types and 13 columns representing a count of graduates for each activity. The Personal Characteristic column is object, and the remaining columns are integers. 

```python
df_chart.describe()
```

```python
update_df_chart = df_chart.drop('Other', inplace=True)
```

```python
df_chart
```

```python
df_chart.to_csv('UG_grad_activities by gender')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (15,15))
df_chart.plot.bar()
plt.title("Undergraduate Activities 2019/0")
plt.xlabel("Activity")
plt.ylabel("Number of Graduates")
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.grid(True, linewidth = 1,  linestyle = "--")
plt.show()
```

```python
plt.savefig('UG_grad_activites_by_gender_2019-0')
```

```python
df_chart
```

```python
females = df_chart.iloc[0]
```

```python
females.plot(kind='bar')
plt.title("Female Undergraduate Graduate Activities 2019/0")
plt.xlabel("Activity")
plt.ylabel("Number of Female Graduates")
plt.grid(True, linewidth = 1,  linestyle = "--")
plt.show()
```

```python
plt.savefig('UG_grad_activites_female_2019-0')
```

```python
males = df_chart.iloc[1]
```

```python
males.plot(kind='bar')
plt.title("Male Undergraduate Graduate Activities 2019/0")
plt.xlabel("Activity")
plt.ylabel("Number of Male Graduates")
plt.grid(True, linewidth = 1,  linestyle = "--")
plt.show()
```

```python
plt.savefig('UG_grad_activites_male_2019-0')
```
